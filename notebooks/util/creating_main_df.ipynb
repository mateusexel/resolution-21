```python
import pandas as pd
import glob
```

```python
def merging_cvm(folder):
    list_files = glob.glob(folder+'/*')
    dfcvm = pd.DataFrame()
    for file in list_files:
        dfcvm = dfcvm.append(pd.read_excel(file))
        
    return dfcvm
```
