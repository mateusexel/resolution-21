```python
import pandas as pd
import numpy as np
import re
import os
import glob
```

```python
def clean_folder(fileDir):
    all_files = glob.glob(fileDir+"/*")
    for file_name in all_files:

        file = open(file_name, 'r+')
        text = file.read()

        text = text.lower()
        text = text.replace('.','')
        
        text = text.replace(')',' ')
        text = text.replace('(',' ')
        text = text.replace('[',' ')
        text = text.replace(']',' ')
        text = text.replace('"', "")
        text = text.replace('', '')
        text = text.replace(';', '')
        text = text.replace(':', '')
        text = text.replace('', '')
        text = text.replace('\uf0b7', '')
        text = text.replace('*', '')
        text = text.replace('', '')
        text = text.replace('', '')
        
        
        
        text = text.replace('\x0c','')
        text = text.replace('\n',' ')
        text = ' '.join(text.split())

        file.seek(0)
        file.truncate(0)
        file.write(text)
        file.close()
        

clean_folder('/media/mateus/F384-E080/fee/txt/')
```
