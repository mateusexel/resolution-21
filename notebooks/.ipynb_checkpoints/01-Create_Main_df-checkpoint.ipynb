## Cleaning Data
### There are many different types of documents, we are only interested on "REGUL FDO"
### we want CNPJ to be cleared from special characters

```python
dfportal=dfportal[dfportal['TP_DOC']=='REGUL FDO']
dfportal['CNPJ_FUNDO']= dfportal['CNPJ_FUNDO'].str.replace('[^0-9]','')
```

### The fund may have sent hundreds of documents, we are only interested on the latest one.

```python
dfportal = dfportal.sort_values(by = ['DT_COMPTC','CNPJ_FUNDO'], ascending=False)
dflatest = dfportal.drop_duplicates('CNPJ_FUNDO', keep='first')
```

```python
dflatest = dflatest[dflatest['DT_COMPTC']>'2015-01-01']
```

### Merging will only happen if both the date the document was published and CNPJ are the same. We will compare them as type string

```python
# dflatest['DT_RECEB']=dflatest['DT_RECEB'].astype(str)
dflatest['CNPJ_FUNDO']=dflatest['CNPJ_FUNDO'].astype(str)
dfthousand['Employer ID\nnumber']=dfthousand['Employer ID\nnumber'].astype(str)
dfthousand['Effective date\n(Admin fee)\ncurrent']=dfthousand['Effective date\n(Admin fee)\ncurrent'].astype(str)
```

```python
dflatest.shape
```

# Merging

```python
eco_and_portal = pd.merge(dflatest, dfthousand, how='inner', left_on=['CNPJ_FUNDO','DT_COMPTC'], right_on=['Employer ID\nnumber','Effective date\n(Admin fee)\ncurrent'])
```

```python
eco_and_portal=eco_and_portal.rename(columns={'Administr % fee\n22Jun20':'fee'})
```

```python
eco_and_portal['fee'] = eco_and_portal['fee'].apply(str)
```

```python
eco_and_portal['fee']=eco_and_portal['fee'].str.replace('.',',')
```
