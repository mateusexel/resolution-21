```python
#converts pdf, returns its text content as a string
def convert(fname, pages=None):
    if not pages:
        pagenums = set()
#         return 'empty PDF'
    else:
        pagenums = set(pages)

    output = io.StringIO()
    manager = PDFResourceManager()
    laparams = LAParams()
    setattr(laparams, 'all_texts', True)
    converter = TextConverter(manager, output, laparams=laparams)
    interpreter = PDFPageInterpreter(manager, converter)
    password="''"

    infile = open(fname, 'rb')
    for page in PDFPage.get_pages(infile, pagenums, password=password):
        interpreter.process_page(page)
    infile.close()
    converter.close()
    text = output.getvalue()
    
    output.close()
    return text 
   
def convertMultiple(pdfDir, txtDir):
    words=''
    n_uniques=[]
    files_list = glob.glob(pdfDir+"/*")
    for pdf in files_list: #iterate through pdfs in pdf directory
        print(pdf)
        text=''
        fileExtension = pdf.split(".")[-1]
        if fileExtension == "pdf":
            pdfFilename = pdf
            
            try:
                text = convert(pdf)

            except PDFTextExtractionNotAllowed as e :
                print(e)
                command = f"qpdf --password={''} --decrypt {pdfFilename} {pdfFilename[:-4]}D.pdf;"
                os.system(command)
                text = convert(pdfFilename[:-4]+'D.pdf')
                
                
        elif fileExtension == "docx":
            pdfFilename=pdfDir + pdf
            document = Document(pdf)
                  
            for para in document.paragraphs: 
                text+=str(para.text)+'\n'
        else:
            next
            
        textFilename = txtDir + pdf.split(".")[-2][46:]+ ".txt"
        textFile = open(textFilename, "w") 
        
        textFile.write(text)
        textFile.close()

pdfDir = "/media/mateus/F384-E080/fee/pdf_administrator/"
txtDir = "/media/mateus/F384-E080/fee/txt_administrator/"
convertMultiple(pdfDir, txtDir)
```
